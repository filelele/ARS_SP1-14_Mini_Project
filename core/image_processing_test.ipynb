```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Hue: Divide by 2 (e.g., 360 → 180).

Saturation: Scale from 0–100% to 0–255 (multiply by 2.55).

Value: Scale from 0–100% to 0–255 (multiply by 2.55).

Normal HSV range

- upper_green = [155,80,30]
- lower_green = [120,30,10]
- upper_white = [80,25,40]
- lower_white = [25,10,30]


```python
lower_green = np.array([90/2,1*2.55,1*2.55])  # Low saturation and brightness for green
upper_green = np.array([165/2,100*2.55,60*2.55])

lower_white = np.array([25/2,0*2.55,30*2.55])  # High brightness for white
upper_white = np.array([80/2,50*2.55,45*2.55])
```

### dilmah

```python
# Load the image
image1 = cv2.imread('sample_to_process\dilmah.jpg')
rgb_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
```

```python
# Convert to HSV color space
hsv_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)

# Create masks for green and white
mask_green1 = cv2.inRange(hsv_image1, lower_green, upper_green)
mask_white1 = cv2.inRange(hsv_image1, lower_white, upper_white)

# Combine masks for the background
background_mask1 = cv2.bitwise_or(mask_green1, mask_white1)
plt.imshow(background_mask1, cmap='gray')
```

```python
# Invert mask to get the object
object_mask1 = cv2.bitwise_not(background_mask1)
plt.imshow(object_mask1, cmap='gray')
```

```python
# Apply morphological operations to clean up noise
kernel_open1 = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
clean_mask1 = cv2.morphologyEx(object_mask1, cv2.MORPH_OPEN, kernel_open1)
kernel_close1 = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
clean_mask1 = cv2.morphologyEx(clean_mask1, cv2.MORPH_CLOSE, kernel_close1)
plt.imshow(clean_mask1, cmap='gray')
```

```python
# Find contours in the mask
contours1, _ = cv2.findContours(clean_mask1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Select the largest contour (assuming it's the object of interest)
if contours1:
    largest_contour1 = max(contours1, key=cv2.contourArea)

    # Get bounding box for the largest contour
    x1, y1, w1, h1 = cv2.boundingRect(largest_contour1)

    # Draw the bounding box on the original image
    cv2.rectangle(image1, (x1, y1), (x1 + w1, y1 + h1), (0, 255, 0), 2)

# Display the result
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
```
